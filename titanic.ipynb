```python
import pandas as pd
import numpy as np
import tensorflow as tf
import sklearn
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
%matplotlib inline
```

I just used the data of training to predict as well as to test .

```python
data1=pd.read_csv("../input/titanic/train.csv")

gender = {'male': 0,'female': 1}
data1.Sex = [gender[item] for item in data1.Sex]
data1.head(20) 
x = np.asanyarray(data1[['Sex']])
y = np.asanyarray(data1[['Survived']])
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
yhat = LR.predict(X_test)
print (classification_report(y_test, yhat))
```
